```python
import numpy as np
import re
```

```python
file = []
with open("logs_cipher.txt","r",encoding='utf8') as f:
    t=""
    for i in f:
        if i == '\n':
            continue
        file.append(i.strip())
```

```python
file
```

```python
ind=[]

pattern = re.compile('Slowly.*')
for i in range(len(file)):
    found = re.search(pattern,file[i])
    if found != None:
        ind.append(i)
```

```python
ind = ind[:-1]
```

```python
len(ind),len(file)
```

```python
ciphertext=[]
for i in ind:
    ciphertext.append(file[i+1])
```

```python
with open("cipher_text.txt","w") as op:
    for c in ciphertext:
        op.write(c+'\n')
```
