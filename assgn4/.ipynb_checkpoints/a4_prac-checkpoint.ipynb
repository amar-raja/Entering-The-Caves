```python
#AND operation
def AND(e1,e2,e3):        
    if (e1=='0' or e2=='0' or e3=='0'):
        return '0'
    else:
        return '1'
    
#XOR opereation
def XOR(e1,e2):                    
    if e1==e2:
        return '0'
    else:
        return '1'
```

```python
def xor_bitwise(e1,e2):            #gives xor value of i/p pairs
    op=''
    for i in range(len(e1)):
        op += XOR(e1[i],e2[i])
    return op
```

```python
def hex2bin(s):
    mp = {'0': "0000",'1': "0001",'2': "0010",'3': "0011",'4': "0100",'5': "0101",'6': "0110",'7': "0111",'8': "1000",'9': "1001",'A': "1010",'B': "1011",'C': "1100",'D': "1101",'E': "1110",'F': "1111"}
    bin = ""
    for i in range(len(s)):
        bin = bin + mp[s[i]]
    return bin

def bin2hex(s):
    mp = {"0000": '0',"0001": '1',"0010": '2',"0011": '3',"0100": '4',"0101": '5',"0110": '6',"0111": '7',"1000": '8',"1001": '9',"1010": 'A',"1011": 'B',"1100": 'C',"1101": 'D',"1110": 'E',"1111": 'F'}
    hex = ""
    for i in range(0, len(s), 4):
        ch = ""
        ch = ch + s[i]
        ch = ch + s[i + 1]
        ch = ch + s[i + 2]
        ch = ch + s[i + 3]
        hex = hex + mp[ch]

    return hex
```

```python
def bin2alpha(s):
    d = {'0000': 'd', '0001': 'e', '0010': 'f', '0011': 'g', '0100': 'h', '0101': 'i', '0110': 'j', '0111': 'k', '1000': 'l', '1001': 'm', '1010': 'n', '1011': 'o', '1100': 'p', '1101': 'q', '1110': 'r', '1111': 's'}
    
    plaintext = ""
    for i in range(0, len(s), 4):
        ch = ""
        ch = ch + s[i]
        ch = ch + s[i + 1]
        ch = ch + s[i + 2]
        ch = ch + s[i + 3]
        plaintext = plaintext + d[ch]

    return plaintext

def alpha2bin(s):
    d = {'d': '0000', 'e': '0001', 'f': '0010', 'g': '0011', 'h': '0100', 'i': '0101', 'j': '0110', 'k': '0111', 'l': '1000', 'm': '1001', 'n': '1010', 'o': '1011', 'p': '1100', 'q': '1101', 'r': '1110', 's': '1111'}
    text = ""
    for i in s:
        text = text + d[i]

    return text
```

```python
def max_key(l):
    key=''
    max=0
    for k,v in l.items():
        if v>max:
            max=v
            key=k
    return key,max
```
