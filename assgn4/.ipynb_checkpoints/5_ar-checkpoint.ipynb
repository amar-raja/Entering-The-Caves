```python
def hex2bin(s):
    mp = {'0': "0000",'1': "0001",'2': "0010",'3': "0011",'4': "0100",'5': "0101",'6': "0110",'7': "0111",'8': "1000",'9': "1001",'A': "1010",'B': "1011",'C': "1100",'D': "1101",'E': "1110",'F': "1111"}
    bin1 = ""
    for i in range(len(s)):
        bin1 = bin1 + mp[s[i]]
    return bin1

def bin2hex(s):
    mp = {"0000": '0',"0001": '1',"0010": '2',"0011": '3',"0100": '4',"0101": '5',"0110": '6',"0111": '7',"1000": '8',"1001": '9',"1010": 'A',"1011": 'B',"1100": 'C',"1101": 'D',"1110": 'E',"1111": 'F'}
    hex = ""
    for i in range(0, len(s), 4):
        ch = ""
        ch = ch + s[i]
        ch = ch + s[i + 1]
        ch = ch + s[i + 2]
        ch = ch + s[i + 3]
        hex = hex + mp[ch]

    return hex
```

```python
# Decimal to binary conversion
def decimal_to_binary(num):
    t = bin(num).replace("0b", "")
    
    if(len(t)%4 != 0):
        div = len(t) / 4
        div = int(div)
        counter =(4 * (div + 1)) - len(t)
        for i in range(0, counter):
            t = '0' + t
    return t
# binary to Decimal conversion
def bin2dec(b):
    res=0
    j=0
    for i in range(len(b)-1,-1,-1):
        res += int(b[i])*(2**(j))
        j+=1
    return res
```

```python
#XOR opereation
def XOR(e1,e2):                    
    if e1==e2:
        return '0'
    else:
        return '1'

def xor_bitwise(e1,e2):            #gives xor value of i/p pairs
    op=''
    for i in range(len(e1)):
        op += XOR(e1[i],e2[i])
    return op
```

```python
 # shifting the bits towards left by nth shifts
def shift_left(b,num):
    s=""
    s = b[num:]
    s = s + b[0:num]
    return s
```

```python
# Dropping parity bits
keyp = [57, 49, 41, 33, 25, 17, 9,
        1, 58, 50, 42, 34, 26, 18,
        10, 2, 59, 51, 43, 35, 27,
        19, 11, 3, 60, 52, 44, 36,
        63, 55, 47, 39, 31, 23, 15,
        7, 62, 54, 46, 38, 30, 22,
        14, 6, 61, 53, 45, 37, 29,
        21, 13, 5, 28, 20, 12, 4 ]
 
#  Number of bit shifts
shift_table = [1, 1, 2, 2,
                2, 2, 2, 2,
                1, 2, 2, 2,
                2, 2, 2, 1 ]
 
# Key- Compression Table : from 56 bits to 48 bits
key_comp = [14, 17, 11, 24, 1, 5,
            3, 28, 15, 6, 21, 10,
            23, 19, 12, 4, 26, 8,
            16, 7, 27, 20, 13, 2,
            41, 52, 31, 37, 47, 55,
            30, 40, 51, 45, 33, 48,
            44, 49, 39, 56, 34, 53,
            46, 42, 50, 36, 29, 32 ]

def key_56to48(k):
    permutation = ""
    for i in range(48):
        permutation = permutation + k[key_comp[i] - 1]
    return permutation
    
```

```python
def encryption(plaintext1,round_key_list):
    plaintext1 = hex2bin(plaintext1)
     
    # Initial Permutation
    plaintext1 = initial_permute(plaintext1)
    # Splitting
    left = plaintext1[0:32]
    right = plaintext1[32:64]
    
    #6ROUND DES
    for i in range(0, 6):
        #  Expansion 
        e = expansion(right)
         
        # XOR RoundKey and expansion output
        xor_op = xor_bitwise(e, round_key_list[i])
        
        #S-BOX output
        s_op = ''
        s_i=0
        for i in range(0,48,6):
            s_op += s_box(xor_op[i:i+6],s_i)
            s_i += 1
        
        #permutation
        op_perm = permutation(s_op)
        
         # XOR left and permutation_output
        result = xor_bitwise(left, op_perm)
        
        #NEXT ROUND inputs
        left, right = right, result
        
    # Combination
    combine = left + right
    
    # Final permutation
    cipher_text = final_permute(combine)
    return cipher_text
```
